## **1. 워드 클라우드(word cloud)**
* 핵심 단어를 시각화하는 기법
* 문서의 키워드, 개념 등을 직관적으로 파악할 수 있게끔 핵심 단어를 시각적으로 돋보이게 하는 기법

```python
text = open('wordData.txt').read()
print(text)
```

```python
from wordcloud import WordCloud
```

```python
# generate() : 단어별 출현 빈도수를 비율로 계산
wordcloud = WordCloud().generate(text)
wordcloud.words_
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (15, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from IPython.display import Image
```

```python
#max_font_size : 최고 출현 빈도가 많은 단어의 크기를 설정 
ordcloud = WordCloud(max_font_size = 50).generate(text)
plt.figure(figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# max_words : 워드 클라우드에 표시되는 단어의 개수를 설정
wordcloud = WordCloud(max_words= 500).generate(text)
plt.figure(figsize=(15, 10))
plt.imshow(wordcloud, interpolation= 'bilinear')
plt.axis('off')
plt.show()
```

```python
text = open('./alice.txt').read()
print(text)
```

```python
from PIL import Image # 이미지에서 마스킹 데이터를 출력
from wordcloud import STOPWORDS # 불용어, 워드 클라우드로 작성할 단어에서 제외할 단어를 지정
import numpy as np
```

```python
alice_mask = np.array(Image.open('./alice_mask.png'))
alice_mask
```

```python
stopwords = set(STOPWORDS)
print(type(stopwords)) # <class 'set'>
stopwords
```

```python
stopwords.add('ha')
stopwords
```

```python
wordcloud = WordCloud(background_color = 'whitesmoke', max_words = 1000, stopwords = stopwords, mask=alice_mask).generate(text)
plt.figure(figsize = (15, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
